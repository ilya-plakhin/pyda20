```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}


def get_owner():
    '''func checks existance and returns owner name from Documens by document number'''
    print('enter doc number:')
    dn = input()
    for doc in documents:
        if doc['number'] == dn:
            print(doc['name'])
            return
    print('doc not found!')
    return


def get_shelf():
    '''func checks existance and returns shelf num from Directories by document number'''
    print('enter doc number:')
    dn = input()
    for shelf, lst in directories.items():
        if dn in lst:
            print(shelf)
    return 'shelf not found!'


def get_list():
    for doc in documents:
        print('number: ' + doc['number'], end=' ')
        print('type: ' + doc['type'], end=' ')
        print('name: ' + doc['name'], end=' ')
        print('shelf: ', end='')
        for shelf, lst in directories.items():
            if doc['number'] in lst:
                print(shelf, end='')
        print()
    return


def add_shelf():
    print('enter shelf number:')
    sh = input()
    for shelf, lst in directories.items():
        if shelf == sh:
            if lst == []:
                print('shelf deleted')
                return
            else:
                print('shelf is not empty!')
                return
    print('shelf not found!')
    return


def del_shelf():
    print('enter shelf number:')
    sh = input()
    for shelf, lst in directories.items():
        if shelf == sh:
            if lst == []:
                directories.pop(shelf)
                print('shelf deleted')
                return
            else:
                print('shelf is not empty!')
                return
    print('shelf not found!')
    return


def decode_command(comm):
    if comm == 'p':
        get_owner()
    elif comm == 's':
        get_shelf()
    elif comm == 'l':
        get_list()
    elif comm == 'as':
        add_shelf()
    elif comm == 'ds':
        del_shelf()
    else:
        print('command unkhown')


print('enter command:')
command = input()
while command != 'q':
    decode_command(command)
    print('enter command:')
    command = input()
print('enough!')
```
